## Checkpoint 3 : Data Visualization + Preprocessing

```python
# Data-preprocessing    
import pandas as pb
df = pb.read_csv('titanic-passengers.csv',sep=',')
df.head()
```

```python
# Data-preprocessing

# getting information on null values 
df.isnull().sum()
```

```python
# Data-preprocessing

# replacing the Nan values with mean : Age
df['Age'].fillna(df['Age'].mean(),inplace=True)

# replacing the Nan values with mode : Cabin
mostfreqCabin = df["Cabin"].mode()[0]
df["Cabin"].fillna(mostfreqCabin,inplace=True)
# replacing the Nan values with mode : Embarked
mostfreqEmbarked = df["Embarked"].mode()[0]
df["Embarked"].fillna(mostfreqEmbarked,inplace=True)

df
```

```python
# Data-preprocessing

# getting information on null values 
df.isnull().sum()
```

```python
#Data visualisation

#Correlation between Sex,Age and Survival
import matplotlib.pyplot as plt
import seaborn as sns 
g=sns.FacetGrid(df,col='Survived', row = 'Sex')
g.map(sns.distplot,'Age')
```

```python
#Data visualisation

#Correlation between Sex,Age and Survival
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.FacetGrid(df,col='Survived', row = 'Embarked')
g.map(sns.distplot,'Pclass')
```

```python
# Data-preprocessing

# labes encoder function :
# Converting the categorial variables to a number : Survived { yes = 1 ;No : 0 }
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Survived'] = encoder.fit_transform(df['Survived'])


# Converting the categorial variables to a number : Sex {male = 1 ; female = 0}
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex']) 

# Converting the categorial variables to a number : Embarked {C=0; Q=1; S=2 }
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Embarked'] = encoder.fit_transform(df['Embarked']) 
```

```python
# Data-Visualization 

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )
# The function provides the correlation between the features. 
# The values given to describe the correlation can be positive or negative. 
# The highest value is 1, and it's only founded on the diagonal
# The idea is, the more the value attributed is close to 1 , 
# the more the correlation between the two features in question is important.
# In our case, the brown color idicates high correlation.
# We can notice that the most important feature, in order to determen ou target (Survived), 
# is the Fare.
```

```python
# Data-Visualization 

#relation between Pclass and survived 
df[["Pclass", "Survived"]].groupby(["Pclass"], as_index=True).mean()
```

```python
# Dropping two columns 
df.drop('Ticket',axis='columns', inplace =True ) 
df.drop('Cabin',axis='columns', inplace =True ) 
# Adding new column 
Title =[]
for i in range (0,891) :
    ch=''
    x=df['Name'][i].index(',')
    y=df['Name'][i].index('.')
    for j in range (x+2,y) :
        ch=ch+df['Name'][i][j]
    Title.append(ch)
df['Title']=Title
df.drop('Name',axis='columns', inplace =True) 
df
```

```python
Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess" : "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
TitleFam=[]
for i in range (0,891) :
    ch=Title_Dictionary[df['Title'][i]]
    TitleFam.append(ch)
df['TitleFam']=TitleFam
df
```

```python
# Converting the categorial variables to a number : Title
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Title'] = encoder.fit_transform(df['Title']) 
# Data-Visualization 

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )
#there is a strong correlation between title, Age and Sex 
```

```python
# Converting the categorial variables to a number : TitleFam
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['TitleFam'] = encoder.fit_transform(df['TitleFam']) 
# Data-Visualization 

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )
#there is a strong correlation between TitleFam, Age and Sex.
#This correlation more obvious 
```

```python
# As the Parch is the number of parents abroad and Sibsp is the number of Siblings/Spouses Aboard
# The FamilySize feature is the sum of the last two features 

df['FamilySize']=df['Parch']+df['SibSp']
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )

# The value of the correlation between FamilySize and Survived is 0.017. 
# We cas conclude that there is not a strong correlation between the feature and the target
```
